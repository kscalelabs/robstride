```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("./c/kinfer_log.ndjson", "r") as f:
    kinfer_log_data = [json.loads(line) for line in f]

kf = pd.DataFrame(kinfer_log_data)
```

```python
joint_amps_matrix = pd.DataFrame(kf["joint_amps"].to_list())
joint_torque = pd.DataFrame(kf["joint_torques"].to_list())
```

```python
joint_amps_matrix 
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(joint_amps_matrix.iloc[:, 0]*0.08)
ax.plot(joint_torque.iloc[:, 0]/2.1*0.08)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(kf["t_us"], joint_torque.iloc[:, 0]/2.1)
plt.show()
```
